```python
import pandas as pd
import numpy as np
import unicodedata
import re
```

```python
def preprocessConsignee(consignee):
    consignee = unicodedata.normalize('NFKD', consignee).encode('ascii', 'ignore').decode().lower()
    regex = re.compile('[^a-z]')
    consignee = regex.sub('', consignee)
    return consignee
```

```python
df = pd.read_csv('LeviStrauss-20160501-20170430.csv', index_col=False)
df.columns = df.columns.str.lower()
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
df['consignee'] = df['consignee'].apply(preprocessConsignee)
df['consignee'].value_counts()
```

```python
countryNames = pd.read_csv('country_names.csv', index_col=0, header=None, names=['consigneeCountry']) # prepared by Dan from consignee adress
countryNames['consigneeCountry'] = countryNames['consigneeCountry'].apply(preprocessConsignee)
countryNames['consigneeCountry'].value_counts()
```

```python
shipperCountry = pd.read_csv('shipper_country.csv', index_col=0, header=None, names=['shipperCountry']) # prepared by Dan from consignee adress
shipperCountry['shipperCountry'] = shipperCountry['shipperCountry'].apply(preprocessConsignee)
shipperCountry['shipperCountry'].value_counts()
```

```python
print(df.shape, countryNames.shape, shipperCountry.shape)
```

```python
df = pd.concat([df, countryNames, shipperCountry], axis=1)
```
