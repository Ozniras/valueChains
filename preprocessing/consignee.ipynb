```python
import pandas as pd
import unicodedata
import re
```

```python
df = pd.read_csv('LeviStrauss-20160501-20170430.csv', index_col=False)
```

```python
df.head(5)
```

```python
df.columns = df.columns.str.lower()
df.dtypes
```

```python
df['consignee'] = df['consignee'].str.lower()
```

```python
def preprocessConsignee(consignee):
    consignee = unicodedata.normalize('NFKD', consignee).encode('ascii', 'ignore').decode()
    regex = re.compile('[^a-zA-Z]')
    consignee = regex.sub('', consignee)
    return consignee
```

```python
df['consignee'].apply(preprocessConsignee)
```
